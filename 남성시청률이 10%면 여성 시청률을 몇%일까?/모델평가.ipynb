# 선형회귀 및 다항회귀 예측 정확도 비교하기

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

# 1. 데이터 준비하기
df = pd.read_csv('제목행번역.csv')
X = df[['남성시청률']]
y = df['여성시청률']
```

## 💡 어려운 용어 쉽게 이해하기

모델이 얼마나 정확한지 판단할 때 쓰는 두 가지 핵심 지표를 초등학생도 이해할 수 있게 설명해줄게요!

### 1. 결정계수 (R-squared)
- **쉽게 말하면?** "모델의 **적중 점수**"예요.
- **비유:** 시험 점수라고 생각하면 쉬워요. 1점에 가까울수록(또는 100%에 가까울수록) 우리가 찍은 선이 데이터 점들을 아주 잘 설명하고 있다는 뜻이에요.
- **판단:** 숫자가 **클수록** 좋습니다! (예: 0.96은 96점짜리 모델이라는 뜻)

### 2. 평균제곱오차 (MSE)
- **쉽게 말하면?** "모델이 **틀린 거리의 평균**"이에요.
- **비유:** 양궁 선수가 과녁 정중앙에서 얼마나 멀리 화살을 쐈는지 거리를 재는 것과 같아요. 실제 점들과 내가 그린 선 사이의 거리가 얼마나 먼지를 계산한 거예요.
- **판단:** 숫자가 **작을수록** 정답에 더 가까운 정확한 모델이에요!

```python
# 2. 선형회귀 (직선 모양) 모델 평가
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

print(f'[선형회귀] 적중 점수(R2): {r2_score(y, y_pred_linear):.4f}')
print(f'[선형회귀] 틀린 거리(MSE): {mean_squared_error(y, y_pred_linear):.4f}')
```

```python
# 3. 다항회귀 (곡선 모양) 모델 평가
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)

print(f'[다항회귀] 적중 점수(R2): {r2_score(y, y_pred_poly):.4f}')
print(f'[다항회귀] 틀린 거리(MSE): {mean_squared_error(y, y_pred_poly):.4f}')
```

### 🏆 최종 결론
다항회귀(곡선) 모델의 **적중 점수가 더 높고**, **틀린 거리(오차)가 더 작기 때문에** 직선 모델보다 더 정확한 모델이라고 할 수 있습니다!

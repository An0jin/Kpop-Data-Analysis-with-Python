# 선형회귀의 실체: 일차 방정식

머신러닝에서 말하는 **선형회귀(Linear Regression)**는 사실 우리가 중학교 수학 시간에 배운 **일차 방정식**과 똑같습니다.

## 핵심 공식
$$y = ax + b$$

- **$y$**: 결과 (예측값)
- **$x$**: 원인 (입력 데이터)
- **$a$**: 기울기 (가중치/Weight)
- **$b$**: y절편 (편향/Bias)

---

## 실전 문제 1: 두 점을 지나는 일차 함수식 구하기

**[주어진 데이터]**
- 데이터 1: $x=1, y=3$
- 데이터 2: $x=3, y=7$

### 풀이 과정
1. **기울기 $a$ 구하기**: $a = \frac{7 - 3}{3 - 1} = 2$
2. **절편 $b$ 구하기**: $3 = 2(1) + b \Rightarrow b = 1$
3. **최종 모델: $y = 2x + 1$**

```python
from sklearn.linear_model import LinearRegression

# 데이터 정의
X = [[1], [3]]
y = [3, 7]

# 모델 학습
model = LinearRegression()
model.fit(X, y)

# 기울기(a)와 절편(b) 추출
a = model.coef_[0]
b = model.intercept_

# 결과 출력
print(f"출력 결과: y = {a:.0f}x + {b:.0f}")
```

## 실전 문제 2: f(x) = 2x + 1 일 때, f(100)의 값은?

학습된 모델(`model`)을 사용하여 x=100일 때의 예측값을 구해봅시다.

```python
# x = 100 대입하여 예측
x_new = [[100]]
prediction = model.predict(x_new)

print(f"f(100)의 예측 결과는: {prediction[0]:.0f}")
```

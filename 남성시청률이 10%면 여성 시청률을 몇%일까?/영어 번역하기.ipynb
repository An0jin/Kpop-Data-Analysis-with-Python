```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_csv('main.csv')

# 2. 컬럼명 번역 매핑 딕셔너리 생성
translation_dict = {
    'K_IDOL_NM': '아이돌이름',
    'K_IDOL_SCCNT_CO': '아이돌검색수',
    'BRDCST_DE': '방송일자',
    'CHNNEL_NM': '채널명',
    'PROGRM_NM': '프로그램명',
    'BRDCST_CO': '방송회차',
    'PROGRM_GENRE_MLSFC_NM': '프로그램장르명',
    'FIXING_CAST_NM': '고정출연자명',
    'CAST_NM': '출연자명',
    'K_IDOL_MALE_WTCHNG_RT': '남성시청률',
    'K_IDOL_FEMALE_WTCHNG_RT': '여성시청률'
}

# 3. 컬럼명 변경
df.rename(columns=translation_dict, inplace=True)

# 4. 결과를 새로운 파일로 저장 (한글 깨짐 방지를 위해 utf-8-sig 지정)
df.to_csv('제목행번역.csv', index=False, encoding='utf-8-sig')
print("번역 완료. '제목행번역.csv'로 저장되었습니다.")
df.head()
```
